```python

x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
u = np.cos(Y)
v = np.sin(X)

fig, ax = plt.subplots()
ax.streamplot(X, Y, u, v)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Streamline plot')

plt.show()
```
